```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
lb.fit_transform(df["Species"])
```

```python
X =df.drop("Species", axis = 1) #Input Variable
y =df["Species"] #Output

```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
```

```python
pred = logreg.predict(x_test)
```

```python
res = pd.DataFrame({"Actual Values " :y_test, "Predicted Values ":pred})
res
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy is ",accuracy_score(pred,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn =KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python
res = pd.DataFrame({"Actual ":y_test, "Predicted ":pred})
res
```

```python
accuracy_score(pred, y_test)
```
